# Santa Fe

## Point mutation

```python
df = pd.read_csv('scripts/out/santafe_pm.csv')
setts = np.array([[int(y) for y in x.split(',')] for x in df['Unnamed: 0']])
df['nodes'] = setts[:, 0]
df['gems'] = setts[:, 1]
df['uses'] = setts[:, 2]
df[['nodes', 'gems', 'uses', 'best fitness', 'mean of best individual', 'std of best indvidiual', 'g_better avg', 'g_worse avg']][((df['uses'] == 0) & (df['gems'] == 0)) |((df['uses'] == 30) & (df['gems'] != 0))]
```
