```python
import pandas as pd
```

```python
df = pd.read_csv('scripts/santa_fe_out/result.csv', index_col=0)
```

## Averages by number of columns
### without gems

```python
df[df.gems == 0].groupby('n_cols').mean()
```

### only with gems

```python
df[df.gems != 0].groupby('n_cols').mean()
```

### all

```python
df.groupby('n_cols').mean()
```

In all cases, the 50 columns is the best in terms of `overall_best` attribute. With exception of 100 nodes, the use of gems brings improvement in terms of `overall_best`.

## Average by number of gems

```python
df.groupby('gems').mean()
```

This only support the findings from above results - `overall_best` is better when using gems, further more mean and std is also lower, indicating more stable convergence.

## Average by number of columns and gems

```python
df.groupby(['n_cols', 'gems']).mean()
```

```python
import seaborn as sns
```

```python
sns.pairplot(df[['gems', 'overall_best']])
```

## Averages by mutation type

```python
df.groupby(['mutation']).mean()
```

```python
df[df.gems != 0].groupby(['mutation', 'strategy']).mean()
```

```python
df[df.gems == 0].groupby(['mutation', 'strategy']).mean()
```

The best combination is probabilistic mutation with match by all strategy. For single mutation, match by active brings significant improvements.

```python
df.groupby(['mutation', 'strategy', 'gems']).mean()
```

Probabilistic mutation with match by active strategy was the only case, where gems did not improved results.
