```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
files = ['scripts/gems0-n_cols10.csv', 'scripts/gems0-n_cols50.csv', 
        'scripts/gems5-n_cols10.csv', 'scripts/gems5-n_cols50.csv']
```

```python
_, ax = plt.subplots(1,1)
for file in files:
    data = pd.DataFrame.from_csv(file)
    data['best_fitness'].plot(ax=ax, label=file)
ax.legend()
ax.set_xlim(0,2000)
```
