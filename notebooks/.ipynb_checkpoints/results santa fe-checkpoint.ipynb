# Santa Fe Trail, results

CSVs created by `scripts/santa_fe.py` script.

```python
import pandas as pd
```

## Load results

```python
df = pd.read_csv('scripts/santa_fe_out/result.csv', index_col=0)
df.head()
```

## Averages by number of columns (without gems)

```python
df[df.gems == 0].groupby('n_cols').mean()
```

## Averages by number of columns (only with gems)

```python
df[df.gems != 0].groupby('n_cols').mean()
```

## Averages by number of columns (all)

```python
df.groupby('n_cols').mean()
```

## Averages by number of gems

```python
df.groupby('gems').mean()
```

## Averages by number of columns and gems

```python
df.groupby(['n_cols', 'gems']).mean()
```

## Averages by mutation type

```python
df.groupby(['mutation', 'strategy']).mean()
```

## Averages by mutation type and gems

```python
df.groupby(['mutation', 'strategy', 'gems']).mean()
```
